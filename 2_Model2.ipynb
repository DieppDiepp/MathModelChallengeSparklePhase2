### I/ Phương trình hấp thụ Glucose từ thức ăn

Phương trình sẽ là:

$$G_{\text{food}}(t) = 
\begin{cases} 
0 & \text{nếu } t < t_0 \\
\frac{G_{\text{max}}}{1 + \exp(-\alpha (t - t_0))} & \text{nếu } t_0 \leq t < t_1 \\
\frac{G_{\text{max}}}{1 + \exp(-\alpha (t - t_0))} \cdot \exp(-\beta (t - t_1)) & \text{nếu } t \geq t_1 
\end{cases}$$


Trong đó

1. $G_{\text{food}}(t)$: Lượng glucose hấp thụ từ thức ăn tại thời điểm $t$.

2. $t_0$: Thời điểm bắt đầu tiêu thụ thức ăn, tức là khi lượng glucose bắt đầu tăng.

3. $t_1$: Thời điểm glucose đạt đỉnh và bắt đầu giảm.

4. $G_{\text{max}}$: Lượng glucose tối đa mà cơ thể hấp thụ được từ thức ăn.

5. $\alpha$: Hệ số điều chỉnh độ dốc của hàm sigmoid trong giai đoạn glucose tăng. Hệ số này điều chỉnh tốc độ tăng của glucose từ thức ăn.

6. $\beta$: Hệ số điều chỉnh tốc độ giảm glucose sau khi đạt đỉnh ở $t_1$. Hệ số này quyết định mức độ nhanh chậm của sự giảm lượng glucose.

7. $\exp$: Hàm số mũ, viết tắt của hàm exponential (lũy thừa), được sử dụng để tạo ra sự thay đổi mượt mà trong quá trình hấp thụ glucose.

### Ý nghĩa từng phần trong công thức:
- Khi $t < t_0$, không có glucose được hấp thụ, vì quá trình ăn uống chưa bắt đầu ($G_{\text{food}}(t) = 0$).
- Khi $t_0 \leq t < t_1$, glucose bắt đầu được hấp thụ vào cơ thể và tăng theo quy luật sigmoid.
- Khi $t \geq t_1$, glucose đã đạt đỉnh và bắt đầu giảm dần theo quy luật mũ, với tốc độ giảm được điều chỉnh bởi $\beta$.

Các đại lượng này mô tả quá trình hấp thụ và tiêu thụ glucose trong cơ thể, phản ánh sự thay đổi lượng glucose theo thời gian khi ăn một bữa ăn.


### Lý do chọn hàm Sigmoid thay vì phương trình vi phân 
- Quá trình hấp thụ glucose từ thức ăn thường có đặc điểm là bắt đầu từ từ, tăng nhanh sau đó chậm lại khi gần đạt đến mức bão hòa. Điều này giống với đặc điểm của hàm sigmoid, nơi tốc độ thay đổi ban đầu thấp, tăng nhanh ở giữa, và sau đó chậm dần khi đạt gần mức tối đa.
- Phương trình vi phân có thể mô tả chính xác những thay đổi tức thời nhưng chưa chắc phản ánh đúng tốc độ thay đổi tổng thể của quá trình hấp thụ theo thời gian.

## II/ Phương trình cho Glucose 
Sự xuất hiện của $G_{\text{food}}(t)$ trong phương trình thay đổi của $G(t)$ là rất quan trọng, vì nó là nguồn cung cấp glucose chính từ thức ăn vào máu. Do đó phương trình mô tả sự thay đổi của nồng độ glucose trong máu ($G(t)$) phải bao gồm thành phần từ thức ăn.

Phương trình sẽ là:

$$
\frac{dG(t)}{dt} = \frac{dG_{\text{food}}(t)}{dt} - k_1 \cdot (G(t) - G_b) - k_2 \cdot I(t) \cdot G(t) + p_8 \cdot [H(t) - H_b]
$$

Trong đó, $\frac{dG_{\text{food}}(t)}{dt}$ đại diện cho tốc độ glucose từ thức ăn hấp thụ vào máu. Điều này giúp mô hình mô tả được sự tăng glucose khi hấp thụ thức ăn và sự đào thải bởi các cơ chế khác.

### Các biến số và tham số:

- **$\frac{dG(t)}{dt}$**: Tốc độ thay đổi nồng độ glucose trong máu.

- **$\frac{dG_{\text{food}}(t)}{dt}$**: Tốc độ hấp thụ glucose từ thức ăn vào máu.

- **$k_1$**: Hằng số tỷ lệ tiêu thụ glucose tự nhiên.

- **$G(t)$**: Nồng độ glucose trong máu tại thời điểm $t$.

- **$G_b$**: Nồng độ glucose cơ bản (mức cân bằng bình thường).

- **$k_2$**: Hằng số tỷ lệ cho thấy hiệu quả của insulin trong việc giảm glucose.

- **$I(t)$**: Nồng độ insulin tại thời điểm $t$.

- **$p_8$**: Hằng số tỷ lệ cho thấy hiệu quả của glucagon trong việc tăng glucose.

- **$H(t)$**: Nồng độ glucagon tại thời điểm $t$.

- **$H_b$**: Nồng độ glucagon cơ bản (mức bình thường).


### Giải thích:
- **$k_2 \cdot I(t) \cdot G(t)$**: Đây là một cách mô tả sự tác động của insulin ($I(t)$) lên việc tiêu thụ glucose ($G(t)$). Ở đây, tác động của insulin tỉ lệ thuận với cả mức insulin hiện có ($I(t)$) và mức glucose hiện tại ($G(t)$). Điều này phản ánh rằng khi có nhiều insulin, nhiều glucose sẽ được tế bào hấp thụ, nhưng tác động đó cũng phụ thuộc vào lượng glucose hiện diện trong máu.

- **$p_8 \cdot (H(t) - H_b)$**: Hàm này mô tả tác động của glucagon ($H(t)$) lên nồng độ glucose trong máu. Khi mức glucagon ($H(t)$) khác với mức cơ bản ($H_b$), glucagon sẽ kích thích gan giải phóng glucose vào máu. Phần $(H(t) - H_b)$ mô tả sự chênh lệch so với mức cơ bản, và $p_8$ là hệ số điều chỉnh tốc độ giải phóng glucose do glucagon.

### Sự khác biệt chính:

1. **Tác động của insulin ($k_2 \cdot I(t) \cdot G(t)$)**: Ở đây, insulin cần có mặt để giúp cơ thể hấp thụ glucose. Quá trình này phụ thuộc vào cả mức insulin lẫn mức glucose. Insulin không tự động làm tăng hoặc giảm glucose nếu không có sẵn glucose trong máu.

2. **Tác động của glucagon ($p_8 \cdot (H(t) - H_b)$)**: Glucagon kích thích gan giải phóng glucose vào máu khi mức glucagon tăng lên so với mức bình thường. Tác động này ít phụ thuộc vào mức glucose hiện tại mà chủ yếu phụ thuộc vào sự thay đổi mức glucagon.

### Vì sao trong phương trình Glucose sự ảnh hưởng của Isulin và Glucagon được xây dựng khác nhau?

- **Insulin**: Tác động phụ thuộc vào cả nồng độ glucose và insulin. Do đó, mô hình cần nhân insulin với glucose ($I(t) \cdot G(t)$) để thể hiện quá trình hấp thụ glucose phụ thuộc vào cả hai yếu tố này.
  
- **Glucagon**: Tác động chỉ phụ thuộc vào sự thay đổi của glucagon so với mức cơ bản. Vì glucagon kích thích việc giải phóng glucose từ gan, sự chênh lệch giữa $H(t)$ và $H_b$ mô tả rõ quá trình này.

Sự khác biệt trong các công thức phản ánh sự khác nhau trong cơ chế hoạt động của hai hormone.

## III/ Phương trình cho Isulin
Hàm insulin được xây dựng để mô tả quá trình thay đổi nồng độ insulin trong máu dựa trên sự biến động của nồng độ glucose. Trong cơ thể, insulin được tiết ra bởi tuyến tụy khi nồng độ glucose tăng cao, nhằm giúp tế bào hấp thụ glucose và giảm nồng độ glucose trong máu. Việc mô hình hóa quá trình này cần phản ánh cả tốc độ tiết insulin khi glucose vượt ngưỡng, và tốc độ phân hủy của insulin khi glucose không còn cao.

### Các thành phần chính để xây dựng hàm insulin:

1. **Phản ứng của insulin với nồng độ glucose**:

     **Khi nồng độ glucose trong máu cao: ($G_{\text{t}}$) > ($G_{\text{b}}$)**
   - Tuyến tụy phát hiện ra nồng độ glucose tăng, dẫn đến việc tiết ra insulin. 
   - Insulin giúp các tế bào (như tế bào cơ và mỡ) hấp thụ glucose từ máu để sử dụng làm năng lượng hoặc lưu trữ dưới dạng glycogen (trong gan). Kết quả là nồng độ glucose trong máu giảm.
   
    **Khi nồng độ glucose trong máu thấp: ($G_{\text{t}}$) < ($G_{\text{b}}$)**
   - Khi glucose được các tế bào hấp thụ, nồng độ glucose trong máu bắt đầu giảm. 
   - Khi mức glucose hạ xuống dưới một ngưỡng nhất định, tuyến tụy sẽ ngừng tiết insulin để tránh giảm quá mức nồng độ glucose.
   - Phản ứng này thường là tuyến tính hoặc theo một hàm phụ thuộc vào sự chênh lệch giữa nồng độ glucose hiện tại ($G(t)$) và ngưỡng ($G_{\text{b}}$).

2. **Phân rã tự nhiên của insulin**:
   - Insulin có thời gian phân hủy ngắn trong máu, nên nó bị phân rã theo thời gian. Do đó, chúng ta cần thêm thành phần mô tả sự giảm nồng độ insulin do phân rã tự nhiên.

### Phương trình mô tả thay đổi của insulin:
Hàm mô tả nồng độ insulin có thể được viết như sau:

$$
\frac{dI(t)}{dt} = \begin{cases}
    i_g \cdot (G(t) - G_{\text{b}}) &  \text{nếu } G(t) > G_{\text{b}} \\
    - i_b \cdot I(t) & \text{nếu } G(t) \leq G_{\text{b}}
\end{cases}
$$

### Giải thích:

1. **Khi $G(t) > G_{\text{b}}$**:
   - Thành phần $i_g \cdot (G(t) - G_{\text{b}})$ đại diện cho sự tiết insulin từ tuyến tụy khi nồng độ glucose vượt ngưỡng $G_{\text{b}}$. $i_g$ là một hệ số điều chỉnh mức độ insulin được tiết ra khi glucose vượt ngưỡng.
   - Thành phần $- i_b \cdot I(t)$ mô tả sự phân rã tự nhiên của insulin theo thời gian. $i_b$ là hệ số phân rã tự nhiên của insulin trong máu. Điều này phản ánh rằng insulin sẽ giảm dần theo thời gian nếu không có thêm tiết insulin.


2. **Khi $G(t) \leq G_{\text{b}}$**:
   - Khi glucose thấp hơn hoặc bằng ngưỡng $G_{\text{b}}$, không có thêm insulin được tiết ra từ tuyến tụy. Do đó, hàm chỉ còn thành phần $- i_b \cdot I(t)$, mô tả sự phân rã tự nhiên của insulin mà không có sự bổ sung.

### Các biến số và tham số:

- $I(t)$: Nồng độ insulin tại thời điểm $t$.
- $G(t)$: Nồng độ glucose tại thời điểm $t$.
- $G_{\text{thresh}}$: Ngưỡng glucose mà tại đó insulin bắt đầu được tiết ra. Đây là giá trị ngưỡng để kích hoạt phản ứng từ tuyến tụy.
- $i_g$: Hệ số điều chỉnh tốc độ tiết insulin khi nồng độ glucose vượt ngưỡng.
- $i_b$: Hệ số phân rã tự nhiên của insulin trong máu.

### Ý nghĩa sinh lý học:
- **Tuyến tụy** tiết insulin khi nồng độ glucose tăng cao để giúp cơ thể hạ mức glucose về trạng thái cân bằng.
- **Phân rã tự nhiên**: Insulin không tồn tại mãi trong máu mà bị phân hủy theo thời gian, cần được bổ sung liên tục nếu nồng độ glucose vẫn cao.


## III/ Phương trình cho Glucagon

Glucagon được tiết ra bởi tuyến tụy khi nồng độ glucose trong máu giảm xuống dưới một mức ngưỡng cụ thể. Mục tiêu của glucagon là tăng nồng độ glucose trong máu bằng cách kích thích gan giải phóng glucose từ glycogen. Việc mô hình hóa quá trình này cần phản ánh cả tốc độ tiết glucagon khi glucose giảm dưới ngưỡng, và tốc độ phân hủy của glucagon khi glucose không còn thấp nữa.

### Mô hình hóa quá trình thay đổi nồng độ glucagon:

1. **Kích thích tiết glucagon (khi $G(t) < G_{\text{b}}$):**
   - Khi nồng độ glucose giảm dưới ngưỡng $G_{\text{b}}$, glucagon được tiết ra để chống lại sự hạ glucose. 
   - Tốc độ tăng nồng độ glucagon tỷ lệ thuận với chênh lệch giữa ngưỡng $G_{\text{b}}$ và nồng độ glucose hiện tại. Đây là cơ chế nhằm tăng cường sản xuất glucose từ gan.

   $$ \frac{dH(t)}{dt} = h_g \cdot [G_{\text{b}} - G(t)] $$

   Trong đó:
   - $H(t)$ là nồng độ glucagon tại thời điểm $t$.
   - $h_g$ là hệ số phản ứng của glucagon khi glucose giảm dưới $G_{\text{b}}$.
   - $G_{\text{b}}$ là ngưỡng glucose dưới đó glucagon bắt đầu tăng.

2. **Giảm nồng độ glucagon (khi $G(t) \geq G_{\text{b}}$):**
   - Khi nồng độ glucose trở về hoặc vượt qua ngưỡng $G_{\text{b}}$, nồng độ glucagon cần được giảm để duy trì sự cân bằng glucose.
   - Tốc độ phân hủy glucagon tỷ lệ thuận với nồng độ glucagon hiện tại, phản ánh sự giảm sản xuất khi glucose đã đủ cao.

   $$ \frac{dH(t)}{dt} = -h_b \cdot H(t) $$

   Trong đó:
   - $h_b$ là hệ số phân hủy glucagon khi nồng độ glucose không còn dưới ngưỡng $G_{\text{b}}$.


Như vậy, mô hình hóa quá trình thay đổi nồng độ glucagon giúp phản ánh cơ chế điều hòa nội tiết của cơ thể để đảm bảo nồng độ glucose trong máu được duy trì trong khoảng bình thường.
Trong mô hình này, phản ứng của glucagon khi glucose giảm dưới ngưỡng $G_{\text{b}}$ được mô tả bởi phương trình vi phân sau:

$$
\frac{dH(t)}{dt} = \begin{cases}
h_g \cdot [G_{\text{b}} - G(t)] & \text{nếu } G(t) < G_{\text{b}} \\
- h_b \cdot H(t) & \text{nếu } G(t) \geq G_{\text{b}}
\end{cases}
$$

Trong đó:

- $H(t)$ là nồng độ của glucagon tại thời điểm $t$.
- $G(t)$ là nồng độ glucose tại thời điểm $t$.
- $G_{\text{b}}$ là ngưỡng glucose dưới đó glucagon bắt đầu tăng.
- $h_g$ là hệ số phản ứng của glucagon khi glucose giảm dưới $G_{\text{b}}$.
- $h_b$ là hệ số phân hủy của glucagon khi nồng độ glucose không còn dưới ngưỡng $G_{\text{b}}$.

### Giải thích:

1. **Khi $G(t) < G_{\text{b}}$:**
   - Glucagon được kích thích tăng lên với tốc độ tỷ lệ thuận với sự chênh lệch giữa ngưỡng $G_{\text{b}}$ và nồng độ glucose hiện tại. Điều này giúp tăng nồng độ glucose trong máu bằng cách kích thích giải phóng glucose từ gan.

2. **Khi $G(t) \geq G_{\text{b}}$:**
   - Glucagon giảm với tốc độ tỷ lệ thuận với nồng độ glucagon hiện tại. Điều này phản ánh cơ chế tự điều chỉnh để giữ nồng độ glucagon ổn định khi nồng độ glucose đã đạt mức bình thường hoặc cao hơn.
   
### Ý nghĩa sinh lý:

- **Kích thích tăng glucagon (khi $G(t) < G_{\text{b}}$):** 
  - Khi glucose giảm quá thấp, cơ thể cần phải tăng nồng độ glucagon để kích thích gan giải phóng glucose và tăng nồng độ glucose trong máu. Điều này giúp duy trì mức glucose cần thiết cho các hoạt động sinh lý và chức năng của tế bào.

- **Giảm glucagon (khi $G(t) \geq G_{\text{b}}$):**
  - Khi nồng độ glucose đã trở về mức bình thường hoặc cao hơn, cần phải giảm nồng độ glucagon để tránh tăng glucose quá mức và duy trì sự cân bằng trong cơ thể.


Mô hình này phản ánh cơ chế điều hòa nội tiết tự nhiên của cơ thể nhằm duy trì nồng độ glucose ổn định và đảm bảo cung cấp đủ năng lượng cho các tế bào.
